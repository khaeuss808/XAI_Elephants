# Rename Input Video Files

Wanted the input videos to have more descriptive titles

```python
from pathlib import Path
```

```python
video_dir = "../data/videos/KTNES_8Aug_tree1"
site_prefix = "KTNES_8Aug_tree1"
```

```python
video_paths = sorted(Path(video_dir).glob("*.AVI"))

print("Preview — files will be renamed as follows:")
for vp in video_paths:
    # extract the 4-digit number from the end of the original name
    number = vp.stem[-4:]
    new_name = f"{site_prefix}_{number}.AVI"
    print(f"  {vp.name}  →  {new_name}")
```

```python
# after confirming above, we can actually rename
for vp in video_paths:
    number = vp.stem[-4:]
    new_name = f"{site_prefix}_{number}.AVI"
    new_path = vp.parent / new_name
    vp.rename(new_path)

print("Done! Files renamed.")
```
